```python
def CommonValues(x, y) :

  result=[]
  for i in range (len(x)) :
    for j in range (len(y)) :
      if (x[i]==y[j]) :
        result.append(y[j])
  return result

array1 = np.array([0, 10, 20, 40, 60])
array2 = np.array([10, 30, 40])

common_values = CommonValues(array1, array2)
print("Common values between two arrays:")
print(common_values)

```
